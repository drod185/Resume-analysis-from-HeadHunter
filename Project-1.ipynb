3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Создаем новый признак 'Опыт работы (месяц)'
def experience_in_months(text):
    """Преобразует текстовое описание опыта работы в количество месяцев."""
    
    # Проверка на пустое значение
    if pd.isna(text) or text == 'Не указано':
        months = np.nan
    else:
        # Берем участок текста с опытом работы
        req_int = text[12 : text.find('  ')]
        
        # Находим первое число в строке
        first_num = int(req_int[:req_int.find(' ')])
        
        # Находим последнее число в строке
        last_space = req_int.rfind(' ')
        second_last_space = req_int.rfind(' ', 0, last_space)
        last_num = int(req_int[second_last_space+1:last_space])
        
        # Находим количество месяцев в зависимости от того, что в строке
        if pd.isna(text) or text == 'Не указано':
            months = np.nan
        if 'лет' in req_int and 'месяц' in req_int:
            months = first_num * 12 + last_num
        elif 'год' in req_int and 'месяц' in req_int:
            months = first_num * 12 + last_num
        elif 'год' in req_int and not 'месяц' in req_int:
            months = first_num * 12
        elif 'лет' in req_int and not 'месяц' in req_int:
            months = first_num * 12
        elif 'месяц' in req_int and not 'год' in req_int and not 'лет' in req_int:
            months = last_num
        else:
            print('Неопознанный промежуток опыта работы')
            print(req_int)
            months = 0
    return months

# Создаем признак 'Опыт работы (месяц)'
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(experience_in_months)
print(hh_df['Опыт работы (месяц)'].median())

# Удаляем признак 'Опыт работы'
hh_df = hh_df.drop('Опыт работы', axis=1)
```

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    * … , готов к переезду , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Создаем признаки 'Город', 'Готовность к переезду', 'Готовность к командировкам'
def sity(text):
    """Вычленяет название города и возвращает нужную категорию"""
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    city_name = text[:text.find(' ,')]
    
    if city_name == 'Москва' or city_name == 'Санкт-Петербург':
        return city_name
    elif city_name in million_cities:
        return 'город-миллинонник'
    else:
        return 'другие'

# Создаем признак 'Город'
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(sity)

# Создаем признак 'Готовность к переезду'
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(
    lambda x: False if 'не готов к перее' in x or 'не готова к перее' in x or 'не хочу перее' in x
    else True)

# Создаем признак 'Готовность к командировкам'
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(
    lambda x: False if 'не готов к командир' in x or 'не готова к командир' in x or 'командировкам' not in x
    else True)

# Удаляем признак 'Опыт работы'
hh_df = hh_df.drop('Город, переезд, командировки', axis=1)

```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Создаем признаки мигалки 'Полная занятость', 'Частичная занятость', 'Проектная работа', 'Стажировка', 'Волонтерство'
hh_df['Полная занятость'] = hh_df['Занятость'].apply(lambda x: True if 'полная занятость' in x else False)
hh_df['Частичная занятость'] = hh_df['Занятость'].apply(lambda x: True if'частичная занятость' in x else False)
hh_df['Проектная работа'] = hh_df['Занятость'].apply(lambda x: True if 'проектная работа' in x else False)
hh_df['Стажировка'] = hh_df['Занятость'].apply(lambda x: True if 'стажировка' in x else False)
hh_df['Волонтерство'] = hh_df['Занятость'].apply(lambda x: True if 'волонтерство' in x else False)

# Создаем признаки мигалки 
hh_df['Полный день'] = hh_df['График'].apply(lambda x: True if 'полный день' in x else False)
hh_df['Сменный график'] = hh_df['График'].apply(lambda x: True if'сменный график' in x else False)
hh_df['Гибкий график'] = hh_df['График'].apply(lambda x: True if 'гибкий график' in x else False)
hh_df['Удаленная работа'] = hh_df['График'].apply(lambda x: True if 'удаленная работа' in x else False)
hh_df['Вахтовый метод'] = hh_df['График'].apply(lambda x: True if 'вахтовый метод' in x else False)

```

Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#Считываем таблицу с курсом валют из файла
table_with_currencies = pd.read_csv('ExchangeRates.csv')

# Переводим в таблицах время в формат datetime
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], format='%d.%m.%Y %H:%M').dt.normalize()
table_with_currencies['date'] = pd.to_datetime(table_with_currencies['date'], format='%d/%m/%y')

# Разбиваем данные в ячейчках столбца ЗП по пробелу, созаем датафрейм из двух столбцов
split_df = hh_df['ЗП'].str.split(expand=True)
split_df.columns = ['Объем валюты', 'Валюта']

# Добвавляем столбцы в наш датафрейм
hh_df = pd.concat([hh_df, split_df], axis=1)
hh_df['Объем валюты'] = hh_df['Объем валюты'].astype(int)

# Делаем карту валют, приводим столбец к ISO
cur_map = {
    'грн.': 'UAH',
    'USD' : 'USD',
    'EUR' : 'EUR',
    'бел.руб.' : 'BYN',
    'KGS' : 'KGS',
    'UZS' : 'UZS',
    'сум' : 'UZS',
    'AZN' : 'AZN',
    'руб.' : 'RUB' 
}
hh_df['Валюта'] = hh_df['Валюта'].replace(cur_map)

# Объединяем две таблицы
hh_df = hh_df.merge(
    table_with_currencies[['currency', 'date', 'close', 'proportion']],
    left_on=['Валюта', 'Обновление резюме'],
    right_on=['currency', 'date'],
    how='left'
)

# Заменяем пустые значения в строках где рубли
hh_df['close'] = hh_df['close'].mask(hh_df['Валюта'] == 'RUB', 1)
hh_df['proportion'] = hh_df['proportion'].mask(hh_df['Валюта'] == 'RUB', 1)

#Создаем столбец ЗП рубли
hh_df['ЗП (руб)'] = hh_df['Объем валюты'] * hh_df['close'] / hh_df['proportion']
#hh_df['ЗП (руб)'] = hh_df['ЗП (руб)'].astype(int)

# Удаляем ненужные столбцы
for_drop = ['ЗП', 'Объем валюты', 'Валюта', 'currency', 'proportion', 'date', 'close']
hh_df = hh_df.drop(columns = for_drop)
```

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15, 4))

log_hh_df = np.log(hh_df['Возраст'] + 1)
histplot = sns.histplot(data=log_hh_df, bins=30, ax=axes)
histplot.set_title('Гистограмма по возрасту')


def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned, mu, lower_bound, upper_bound

outliers, cleaned, mu, lower_bound, upper_bound  = outliers_z_score(hh_df, 'Возраст', log_scale=True)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
print(hh_df['Возраст'].mean(), lower_bound, upper_bound)
histplot.axvline(lower_bound, color='k', lw=2)
histplot.axvline(upper_bound, color='k', lw=2)
histplot.axvline(mu, color='k', lw=2)

plt.show()
print(outliers['Возраст'])
```
